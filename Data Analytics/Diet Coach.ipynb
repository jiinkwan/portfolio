### 6. Adding new information

BMI is commonly used in masuring body composition even though there are criticism that it simplifies it too much. However for the sake of simplicity, we are going to use this to identifies those two groups that we wanted to identify.

The new columns to be added are:

1. Weight Loss(goal): Initial Weight - Target Weight
2. Weight Loss(current): Initial Weight - Lowest Weight
3. BMI: Weight / (height(meter) / height(meter))

````python
data['Weight Loss(goal)'] = data['Initial Weight'] - data['Target Weight']
data['Weight Loss(current)'] = data['Initial Weight'] - data['Lowest Weight']
data['BMI'] = data['Initial Weight'] /(data['Height(clean)']/100)**2

data[['Name','Weight Loss(goal)', 'Weight Loss(current)', 'BMI']].head()
````

### 7. Identifying users who entered invalid data

Since we are only considering about paid users, we are going to exclude unpaid users which we can identify it by Status. Paid users' status is completed

The rules we are going to use to identify them are:
1. NaN in Age, Height, Initial Weight, Lowest Weight, or Target Weight
2. Too low (less than 140cm) or high (more than 200)height
3. Too low (less than 18.5) or too high (30.0) BMI
4. Weight Loss(Goal) negative (usually people want to loss weight not gain)

````python
def find_invalid_user(row):
    statusCompleted = row['Status']
    age = row['Age(clean)']
    height = row['Height(clean)']
    initialWeight = row['Initial Weight']
    lowerstWeight = row['Lowest Weight']
    targetWeight = row['Target Weight']
    bmi = row['BMI']
    weightLossGoal = row['Weight Loss(goal)']
    if statusCompleted != 'completed':
        return False    
    if np.isnan(age) | np.isnan(height) | np.isnan(initialWeight) | np.isnan(lowerstWeight)|np.isnan(targetWeight):
        return True
    if (height > 200) | (height < 140):
        return True
    if np.isnan(bmi) | (bmi > 30) | (bmi < 18.5):
        return True
    if weightLossGoal < 0:
        return True
    return False
````

````python
# From the following we can easily check if the logic is correct
dummy = {
    'Status': 'completed',
    'Age(clean)' : 20,
    'Height(clean)' : 201,
    'Initial Weight': 150,
    'Lowest Weight': 100,
    'Target Weight': 100,
    'BMI': 20,
    'Weight Loss(goal)': 10
}

find_invalid_user(dummy)
````

````python
data["Invalid"] = data.apply(find_invalid_user,axis = 1)
data[data['Invalid'] == True].shape
````

````python
data.loc[data['Invalid'] == True, 'Name']
````

We can send notification to the users who are indenfitied as entered invalid data.

### VIP Check

To find the VIP users we are going use the following rule.

1. Status == "completed" (paid user)
2. Weight Loss(goal), Weight Loss(current), BMI not NaN
3. Weight Loss(current) more than 10kg
4. BMI > 30
5. Weight Loss(Current) > Weight Loss(goal) (who met their diet goal)

````python
def checkVIP(row):
    status = row['Status']
    weightLossGoal = row['Weight Loss(goal)']
    weightLossCurrent = row['Weight Loss(current)']
    bmi = row['BMI']
    if status != 'completed':
        return False
    if np.isnan(weightLossGoal) | np.isnan(weightLossCurrent) | np.isnan(bmi):
        return False
    if weightLossCurrent < 10:
        return False
    if bmi < 30:
        return False
    if weightLossCurrent < weightLossGoal:
        return False
    return True
````

````python
dummy = {
    'Status': 'completed',
    'Weight Loss(goal)': 15,
    'Weight Loss(current)': 10,
    'BMI': 35
}
checkVIP(dummy)
````

````python
data['VIP']=data.apply(checkVIP, axis = 1)
````

````python
data.loc[data['VIP']==True, "Name"]
````

Those users could be VIP users and be targeted with promotions.

### Paid, cancelled, refunded ratio by Gender and Age

First we are going to age group.

1. Under 17
2. 18 ~ 24
3. 25 ~ 35
4. 36 ~ 44
5. 45 ~ 54
6. Over 55

````python
def cutAgeGroup(x):
    if x <= 17:
        return '00 ~ 17'
    elif 17 < x <= 24:
        return '18 ~ 24'
    elif 24 < x <= 35:
        return '25 ~ 35'
    elif 35 < x <= 44:
        return '36 ~ 44'
    elif 44 < x <= 54:
        return '45 ~ 54'
    elif 54 < x:
        return '55 ~ 99'
    else:
        return np.nan
````

````python
data['Age(group)'] = data['Age(clean)'].apply(cutAgeGroup)
````

````python
columns = ['Name', 'Age(clean)', 'Height(clean)', 'Initial Weight',
       'Lowest Weight', 'Target Weight', 'Status', 'Weight Loss(goal)',
       'Weight Loss(current)', 'BMI', 'Invalid', 'VIP', 'Age(group)',
       'Gender(clean)','Purchased At', 'Channel']
data = data[columns]
````

### Conversion Rate by Coach

This could be useful and very sensitive too as it shows the coaches performance directly. However it's important to know that who are doing well and who are not. If we can provide better training to underperformed one we can increase the performance or bench mark overperformed ones. 

````python
coach = pd.read_csv("data/diet/coach.csv", index_col="Access Code")
print(coach.shape)
coach.head()
````

````python
## Merge user and coach datasets
match = pd.merge(data[['Name','Status']],coach,how = 'inner', left_index = True, right_index = True)
match.head()
````

````python
coachOrder = match.columns[2:102]
matchPivoted = pd.pivot_table(data = match, index = 'Status', values = coachOrder, aggfunc = 'sum')
````

### Conversion / Cancellation rate

Finally conversion and cancellation rate. If a coach has high conversion rate, he or she is doing something well, if a coach has high cancellation rate, he or she is not doing so well. We will find the top performing ones and bottom performing ones, so that we can give right feedback. To be fair, who has less than 100 trainees, we will exclude from the result.

````python
matchPivotedT = matchPivoted.T.reindex(coachOrder)
````

````python
matchPivotedT['total'] = matchPivotedT['cancelled'] + matchPivotedT['completed'] +matchPivotedT['refunded']
matchPivotedT['conversion_rate'] = matchPivotedT['completed'] / matchPivotedT['total']
matchPivotedT['cancellation rate'] = (matchPivotedT['cancelled'] + matchPivotedT['refunded']) / matchPivotedT['total']
matchPivotedT
````

````python
matchPivotedT[matchPivotedT['total'] >= 100].sort_values(by ='conversion_rate', ascending=False).head()
````

````python
matchPivotedT[matchPivotedT['total'] >= 100].sort_values(by ='cancellation rate', ascending=False).head()
````
