# Predicting House Sale Prices

In this project, we are going to work with housing data for the city of Ames, Iowa, United States from 2006 to 2010. You can read more about why the data was collected [here](https://doi.org/10.1080/10691898.2011.11889627). You can also read about the different columns in the data [here](https://s3.amazonaws.com/dq-content/307/data_description.txt)

We are going to start by setting up a pipeline of functions that will let us quickly iterate on different models. 

train => trainform_features() => select_features() => train_and_test()

## Setting up the environment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('AmesHousing.tsv', delimiter='\t')
```

```python
def transform_features(df):
    return df

def select_features(df):
    return df[['Gr Liv Area','SalePrice']]
    
def train_and_test(df):
    lr = LinearRegression()

    train = df.iloc[0:1460]
    test = df.iloc[1460:]

    #numeric_train = select_features(train).select_dtypes(include = ['integer','float'])
    #numeric_test = select_features(test).select_dtypes(include = ['integer','float'])
    numeric_train = train.select_dtypes(include = ['integer','float'])
    numeric_test = test.select_dtypes(include = ['integer','float'])
    
    features = numeric_train.columns.drop('SalePrice')
    target = 'SalePrice'

    lr.fit(numeric_train[features], numeric_train[target])
    predictions = lr.predict(numeric_test[features])

    mse = mean_squared_error(numeric_test[target],predictions)
    rmse = np.sqrt(mse)
    return rmse

tranformed_df = transform_features(df)
selected_df = select_features(tranformed_df)
rmse = train_and_test(selected_df)

rmse
```

## Feature Engineering

This is experiment part before applying it to the tranform_features fuction.

The goal of this function is to:
    
    - remove features that we don't want to use in the model, just based on the number of missing values or leakage
    - transform features into the proper format (numerical to categorical, scaling numerical, filling in missing values, etc)
    - create new featuers by combining other features
 

### Null Values Ratio

Drop any column column that more than 25% of missing values


```python
null_value_columns =  df.isnull().sum()
null_value_ratio = null_value_columns[null_value_columns != 0]/df.shape[0]
null_value_high = null_value_ratio[null_value_ratio > 0.25].index
df.drop(null_value_high, axis = 1, inplace = True)
```

Fill missing values with the most popular value in the column if the column has less than 5% of missing values. Now we have 9 columns that have less than 5% of missing values but still have missing values. 

```python
numeric_df = df.select_dtypes(include = ['integer','float'])
numeric_null_ratio = numeric_df.isnull().sum() / len(df)
numeric_missing_under_5 = numeric_null_ratio[(numeric_null_ratio < 0.05) & (numeric_null_ratio != 0)]
numeric_missing_under_5_index = numeric_missing_under_5.index
for i in numeric_missing_under_5_index:
    df[i].fillna(df[i].mode()[0],inplace = True)
```

There are still the following missing values. Let's not consider them in the model for **now** because there aren't clear benefit for using those parameters. 

#### New Features

How long ago the property was remodeled when it was sold?   
How long did it take to remodel?  
How old was the property when it was sold?  

```python
df['years_remod_before_sale']= df['Yr Sold']-df['Year Remod/Add']
df['years_until_remod']= df['Year Remod/Add']-df['Year Built']
df['years_until_sale'] = df['Yr Sold']-df['Year Built']
```

```python
df['years_remod_before_sale'].value_counts().sort_values()
```

This is weird that some of the years_remod_before_sale values came back negative. That means either they were remodeled after sales, in which case it is irrelevent to the final price, or human error. Either way let's remove those records because it will distort the model and they are only few.

```python
df.drop(df[df['years_remod_before_sale'] < 0].index, inplace = True)
df.drop(df[df['years_until_remod'] < 0].index, inplace = True) # same with the previous one
df.drop(df[df['years_until_sale'] < 0].index, inplace = True)
```

For the simplicity, we are going to only use the features that are numeric for now. 'Yr Sold', 'Mo Sold', 'Sale Type', and 'Sale Condition' are going be dropped because they can potentially leak data about the final sale.  

'Order' and 'PID' are also going to be dropped because they identifier, they don't really help the model. 

'Year Built' and 'Year Remod/Add' are also to be dropped because they are already used to create a new features. 

Let's update the functions and see how our model has been improved. 

```python
df = pd.read_csv('AmesHousing.tsv', delimiter='\t')

def transform_features(df):
    # High null value ratio columns dropped
    null_value_columns =  df.isnull().sum()
    null_value_ratio = null_value_columns[null_value_columns != 0]/df.shape[0]
    null_value_high = null_value_ratio[null_value_ratio > 0.25].index
    df.drop(null_value_high, axis = 1, inplace = True)
    
    # Imputing missing values
    numeric_df = df.select_dtypes(include = ['integer','float'])
    numeric_null_ratio = numeric_df.isnull().sum() / len(df)
    numeric_missing_under_5 = numeric_null_ratio[(numeric_null_ratio < 0.05) & (numeric_null_ratio != 0)]
    numeric_missing_under_5_index = numeric_missing_under_5.index
    for i in numeric_missing_under_5_index:
        df[i].fillna(df[i].mode()[0],inplace = True)
    
    # How long ago the property was remodeled when it was sold?
    # How long did it take to remodel?
    # How old was the property when it was sold?
    df['years_remod_before_sale']= df['Yr Sold']-df['Year Remod/Add']
    df['years_until_remod']= df['Year Remod/Add']-df['Year Built']
    df['years_until_sale'] = df['Yr Sold']-df['Year Built']
    
    df.drop(df[df['years_remod_before_sale'] < 0].index, inplace = True)
    df.drop(df[df['years_until_remod'] < 0].index, inplace = True) # same with the previous one
    df.drop(df[df['years_until_sale'] < 0].index, inplace = True)

    drop_columns = ['Order','PID','Yr Sold', 'Mo Sold', 'Sale Type'
                    , 'Sale Condition','Year Built', 'Year Remod/Add']
    
    df.drop(drop_columns, axis = 1, inplace = True)
    
    return df

def select_features(df):
    return df[['Gr Liv Area','SalePrice']]
    
def train_and_test(df):
    lr = LinearRegression()

    train = df.iloc[0:1460]
    test = df.iloc[1460:]

    #numeric_train = select_features(train).select_dtypes(include = ['integer','float'])
    #numeric_test = select_features(test).select_dtypes(include = ['integer','float'])
    numeric_train = train.select_dtypes(include = ['integer','float'])
    numeric_test = test.select_dtypes(include = ['integer','float'])
    
    features = numeric_train.columns.drop('SalePrice')
    target = 'SalePrice'

    lr.fit(numeric_train[features], numeric_train[target])
    predictions = lr.predict(numeric_test[features])

    mse = mean_squared_error(numeric_test[target],predictions)
    rmse = np.sqrt(mse)
    return rmse

tranformed_df = transform_features(df)
selected_df = select_features(tranformed_df)
rmse = train_and_test(selected_df)

rmse
```

RMSE score improved by about 1500.

- Total Bsmt SF and 1st Flr SF
- TotRms AbvGrd and Gr Liv Area
- Garage Cars and Garage Area

It might be a good idea to drop one of the correlated columns. 

```python
coef_saleprice[high_coef].index
```

```python
# Drop 'Total Bsmt SF', it's highly correlated to 1st Flr SF and not all house have basements even 
# though it's pretty standard in midwest. 
# Drop TotRms AbvGrd because it is correlated to Gr Liv Area and Gr Liv Area is easier to handle. 
# Drop Garage Cars for the same reason. 
numeric_features = coef_saleprice[high_coef].index.drop(
        ['Total Bsmt SF','TotRms AbvGrd','Garage Cars','Garage Yr Blt'])
```

Now we are left with the folloiwng numeric features. Let's test how they are doing.

There are indeed some different on average price depending on the lot shape and config. They can't be used directly in the model because linear regression model only takes numerical values as features. Let's turn them into dummy columns.
