![alt text](https://s3.amazonaws.com/dq-content/191/chinook-schema.svg "Chinook Schema")  

The database retains 'state', so if we run a qeury with a CREATE or DROP twice, the query will fail.

````python
import sqlite3
import pandas as pd
import matplotlib
%matplotlib inline
````

- We'll create some helper functions in python to save some time.
- We'll use a context manager to handle the connection to the SQLite database.

```python
with sqlite3.connect('chinook.db') as conn:
    # we can now access `conn`
    # inside an indented block
    query_result = pd.read_sql(my_query, conn)
```
With a few extra lines, we can turn this code into a function:

````python
def run_query(q):
    with sqlite3.connect('chinook.db') as conn:
        return pd.read_sql(q, conn)
````

```python
with sqlite3.connect('chinook.db') as conn:
    conn.isolation_level = None
    conn.execute(c)
```
The ```conn.isolation_level = None``` line above tells SQLite to autocommit any changes.

````python
def run_command(c):
    with sqlite3.connect('chinook.db') as conn:
        return conn.execute(c)
````

````python
def show_tables():
    with sqlite3.connect('chinook.db') as conn:
            q = '''SELECT
                        name,
                        type
                    FROM sqlite_master
                    WHERE type IN ("table","view");'''
            return pd.read_sql(q, conn)
````

````python
show_tables()
````

## Selecting Albums to Purchase

The Chinook record store has just signed a deal with a new record label, and you've been tasked with selecting the first three albums that will be added to the store, from a list of four. All four albums are by artists that don't have any tracks in the store right now - we have the artist names, and the genre of music they produce:

|Artist Name |	Genre
|---|---|
|Regal |	Hip-Hop
|Red Tone |	Punk
|Meteor and the Girls |	Pop
|Slim Jim Bites |	Blues

The record label specializes in artists from the USA, and they have given Chinook some money to advertise the new albums in the USA, so we're interested in finding out which genres sell the best in the USA.

### Numbe of tracks sold in The USA by genres

````python
genre_query = '''
    SELECT 
            *
    FROM genre;
        
    
    '''

run_query(genre_query)
````

````python
invoice_line_query = '''
    SELECT 
            *
    FROM invoice;
    '''

run_query(invoice_line_query).shape
````

````python
invoice_line_query = '''
    SELECT 
            *
    FROM invoice_line
    WHERE quantity != 1;
    '''

run_query(invoice_line_query)
````

````python
track_query = '''
    SELECT 
            *
    FROM track;
    '''

run_query(track_query).shape
````

Because there's no invoice line of which quantity is not eqault to 1, we can use count instead of sum.

````python
q  = '''
    SELECT 
            g.name genre_name, count(il.quantity) no_track_sold, count(il.quantity) / (
                 SELECT CAST(COUNT(*) AS FLOAT) from invoice_line il 
                 INNER JOIN invoice i ON i.invoice_id = il.invoice_id
                 WHERE i.billing_country = 'USA' ) ratio
    FROM invoice_line il 
         INNER JOIN invoice i ON i.invoice_id = il.invoice_id 
         INNER JOIN track t ON t.track_id = il.track_id
         INNER JOIN genre g ON g.genre_id = t.genre_id
    WHERE i.billing_country = 'USA'    
    GROUP BY 1;
        
    
    '''

run_query(q)
````

````python
run_query(q)[['genre_name','no_track_sold']].plot.bar(x = 'genre_name')
````

By far, Rock is the most popular genre in the USA. But, none of the artist that we are considering is in Rock. Let's narrow our fields to those genres. 

|Artist Name |	Genre
|---|---|
|Regal |	Hip-Hop
|Red Tone |	Punk
|Meteor and the Girls |	Pop
|Slim Jim Bites |	Blues

````python
q  = '''
    SELECT 
            g.name genre_name, count(il.quantity) no_track_sold, count(il.quantity) / (
                 SELECT CAST(COUNT(*) AS FLOAT) from invoice_line il 
                 INNER JOIN invoice i ON i.invoice_id = il.invoice_id
                 WHERE i.billing_country = 'USA' ) ratio
    FROM invoice_line il 
         INNER JOIN invoice i ON i.invoice_id = il.invoice_id 
         INNER JOIN track t ON t.track_id = il.track_id
         INNER JOIN genre g ON g.genre_id = t.genre_id
    WHERE i.billing_country = 'USA' 
    AND g.name in ('Hip Hop/Rap','Punk','Pop','Blues', 'Alternative & Punk')
    GROUP BY 1;
        
    
    '''

run_query(q)

run_query(q)[['genre_name','no_track_sold']].sort_values(by='no_track_sold', ascending = False).plot.bar(x = 'genre_name')
````

Among those 4 genres, Punk, Blues, Pop are rather popular ones.

## Customer Analysis

The next task is to analyze the sales data for customers from each different country. 

- total number of customers
- total value of sales
- average value of sales per customer
- average order value

````python
c = '''DROP VIEW number_of_customer'''
run_command(c)
````

````python
c = '''
    CREATE VIEW number_of_customer AS
    WITH one_cust_country AS (
        SELECT country, COUNT(country) as cc
        FROM customer
        GROUP BY 1
        HAVING cc = 1

    )
    SELECT CASE WHEN country in (SELECT country FROM one_cust_country) THEN 'Other'
           ELSE country END country,
           COUNT(customer_id) no_of_customer
    FROM CUSTOMER
    GROUP BY 1
    ORDER BY 2 DESC
 '''

run_command(c)
````

````python
q = '''
    SELECT
        country,
        no_of_customer
    FROM
        (
        SELECT
            s.*,
            CASE
                WHEN s.country = "Other" THEN 1
                ELSE 0
            END AS sort
        FROM number_of_customer s
       )
    ORDER BY sort ASC
    '''

number_of_customer = run_query(q)
````

````python
c  = '''
    CREATE VIEW sales_by_country_all AS
    WITH one_cust_country AS (
        SELECT country, COUNT(country) as cc
        FROM customer
        GROUP BY 1
        HAVING cc = 1

    )
    
    SELECT 
            CASE 
                WHEN c.country in (SELECT a.country FROM one_cust_country a) THEN 'Other'
                ELSE c.country END country,
            sum(i.total) as total
    FROM invoice_line il 
         INNER JOIN invoice i ON i.invoice_id = il.invoice_id 
         INNER JOIN track t ON t.track_id = il.track_id
         INNER JOIN genre g ON g.genre_id = t.genre_id
         INNER JOIN customer c ON c.customer_id = i.customer_id
         INNER JOIN employee e ON e.employee_id = c.support_rep_id
    GROUP BY 1
    ORDER BY 2 DESC;
    '''
run_command(c)
````

````python
q = '''
    SELECT
        country,
        total
    FROM
        (
        SELECT
            s.*,
            CASE
                WHEN s.country = "Other" THEN 1
                ELSE 0
            END AS sort
        FROM sales_by_country_all s
       )
    ORDER BY sort ASC
    '''

sales_by_country = run_query(q)
````

#### Average Value of Sales per Customer

````python
c  = '''
    CREATE VIEW avg_sales_by_country_all AS
    WITH one_cust_country AS (
        SELECT country, COUNT(country) as cc
        FROM customer
        GROUP BY 1
        HAVING cc = 1

    )
    
    SELECT 
            CASE 
                WHEN c.country in (SELECT a.country FROM one_cust_country a) THEN 'Other'
                ELSE c.country END country,
            sum(i.total) / count(distinct c.customer_id) as avg_per_customer
    FROM invoice_line il 
         INNER JOIN invoice i ON i.invoice_id = il.invoice_id 
         INNER JOIN track t ON t.track_id = il.track_id
         INNER JOIN genre g ON g.genre_id = t.genre_id
         INNER JOIN customer c ON c.customer_id = i.customer_id
         INNER JOIN employee e ON e.employee_id = c.support_rep_id
    GROUP BY 1
    ORDER BY 2 DESC;
    '''
run_command(c)
#run_query(c)
````

````python
q = '''
    SELECT
        country,
        avg_per_customer
    FROM
        (
        SELECT
            s.*,
            CASE
                WHEN s.country = "Other" THEN 1
                ELSE 0
            END AS sort
        FROM avg_sales_by_country_all s
       )
    ORDER BY sort ASC
    '''

avg_sales_per_customer = run_query(q)
````

#### Average Order Value

````python
c = '''
    CREATE VIEW averag_order_value AS
    WITH one_cust_country AS (
        SELECT country, COUNT(country) as cc
        FROM customer
        GROUP BY 1
        HAVING cc = 1

    )
    
    SELECT 
            CASE 
                WHEN c.country in (SELECT a.country FROM one_cust_country a) THEN 'Other'
                ELSE c.country END country,
            sum(i.total) / count(distinct i.invoice_id) as avg_order_value
    FROM invoice_line il 
         INNER JOIN invoice i ON i.invoice_id = il.invoice_id 
         INNER JOIN track t ON t.track_id = il.track_id
         INNER JOIN genre g ON g.genre_id = t.genre_id
         INNER JOIN customer c ON c.customer_id = i.customer_id
         INNER JOIN employee e ON e.employee_id = c.support_rep_id
    GROUP BY 1
    ORDER BY 2 DESC;
    '''
run_command(c)
#run_query(c)
````

````python
q = '''
    SELECT
        country,
        avg_order_value
    FROM
        (
        SELECT
            s.*,
            CASE
                WHEN s.country = "Other" THEN 1
                ELSE 0
            END AS sort
        FROM averag_order_value s
       )
    ORDER BY sort ASC
    '''

avg_order_value = run_query(q)
````

### Albums vs Individual Track Purchase

The Chinook store is setup in a way that allows customer to make purchases in one of the two ways:

- purchase a whole album
- purchase a collection of one or more individual tracks.

The store does not let customers purchase a whole album, and then add individual tracks to that same purchase (unless they do that by choosing each track manually). When customers purchase albums they are charged the same price as if they had purchased each of those tracks separately.

````python
q = '''
    SELECT COUNT(*)
    FROM invoice i JOIN invoice_line il ON il.invoice_id = i.invoice_id
    '''
````

````python
run_query(q)
````

````python
q  = '''
    WITH album_track AS(
        SELECT 
                a.album_id,
                a.title,
                t.track_id
        FROM invoice_line il 
             INNER JOIN invoice i ON i.invoice_id = il.invoice_id 
             INNER JOIN track t ON t.track_id = il.track_id
             INNER JOIN genre g ON g.genre_id = t.genre_id
             INNER JOIN customer c ON c.customer_id = i.customer_id
             INNER JOIN employee e ON e.employee_id = c.support_rep_id
             INNER JOIN album a ON a.album_id = t.album_id
        
    ),
    puchase_track AS(
        SELECT 
                i.invoice_id,
                a.title,
                t.track_id
        FROM invoice_line il 
             INNER JOIN invoice i ON i.invoice_id = il.invoice_id 
             INNER JOIN track t ON t.track_id = il.track_id
             INNER JOIN genre g ON g.genre_id = t.genre_id
             INNER JOIN customer c ON c.customer_id = i.customer_id
             INNER JOIN employee e ON e.employee_id = c.support_rep_id
             INNER JOIN album a ON a.album_id = t.album_id
    )
    
    select * from puchase_track pt JOIN album_track at ON at.track_id = pt.track_id
    
        
    
    '''

run_query(q)
````
