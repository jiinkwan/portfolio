## 전처리 (Preprocessing)

### Encode Sex

```python
train.loc[train["Sex"] == "male", "Sex_encode"] = 0
train.loc[train["Sex"] == "female", "Sex_encode"] = 1
print(train.shape)
train[["Sex", "Sex_encode"]].head()
```

```python
test.loc[test["Sex"] == "male", "Sex_encode"] = 0
test.loc[test["Sex"] == "female", "Sex_encode"] = 1
print(test.shape)
test[["Sex", "Sex_encode"]].head()
```

## Fill in missing Fare

```python
train[train["Fare"].isnull()]
```

```python
test[test["Fare"].isnull()]
```

```python
test.loc[test["Fare"].isnull(), "Fare"] = 0 # test data의 Null값을 0으로 대체

test[test["Fare"].isnull()]
```

## Encode Embarked 

```python
# C == 0, S == 1, Q == 2 로 인코딩 시 
# 2 * S == Q?
# S + S == Q? 로 판단하는 오류가 생길 수 있다.

# One Hot Encoding을 사용 해야 한다
# C == [1, 0, 0]
# S == [0, 1, 0]
# Q == [0, 0, 1]
# C == [True, False, False]
# S == [False, True, False]
# Q == [False, False, True]

train["Embarked_C"] = train["Embarked"] == "C"
train["Embarked_S"] = train["Embarked"] == "S"
train["Embarked_Q"] = train["Embarked"] == "Q"
print(train.shape)
train[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
# C == 0, S == 1, Q == 2 로 인코딩 시 
# 2 * S == Q?
# S + S == Q? 로 판단하는 오류가 생길 수 있다.

# One Hot Encoding을 사용 해야 한다
# C == [1, 0, 0]
# S == [0, 1, 0]
# Q == [0, 0, 1]
# C == [True, False, False]
# S == [False, True, False]
# Q == [False, False, True]

test["Embarked_C"] = test["Embarked"] == "C"
test["Embarked_S"] = test["Embarked"] == "S"
test["Embarked_Q"] = test["Embarked"] == "Q"
print(test.shape)
test[["Embarked", "Embarked_C", "Embarked_S", "Embarked_Q"]].head()
```

```python
feature_names = ["Pclass","Sex_encode", "Fare", "Embarked_C","Embarked_S", "Embarked_Q"] # Decision Tree의 feature로 들어갈 리스트
feature_names
```

```python
X_train = train[feature_names] # 변수로 뽑아낸 data frame 저장
print(X_train.shape)
X_train.head()
```

```python
X_test = test[feature_names] # 변수로 뽑아낸 data frame 저장
print(X_test.shape)
X_test.head()
```
