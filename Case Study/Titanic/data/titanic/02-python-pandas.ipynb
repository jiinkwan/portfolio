## Column 추가 & 수정

```python
train["Category"] = "Titanic" # 컬럼 추가, 값은 모두 같음
```

```python
train.head()
```

```python
train["Id"] = range(891) #컬럼 추가, 값은 모두 다름
# pandas 컬럼 추가, 행 수만 맞으면 일단 들어감.
train.head()
```

```python
train["familysize"] = train["SibSp"] + train["Parch"] + 1
train.head()
```

```python
# 조건마다 추가
train["France"] = train["Embarked"] == "C"
train["UK"] = train["Embarked"] != "C"

train[["Embarked","France","UK"]].head() # 조건마다 컬럼이 추가 해야 되기 때문에 양이 많아지면 관리 힘듦
```

```python
train.loc[train["Embarked"] == "C", "Nationality"] = "France"
train.loc[train["Embarked"] != "C", "Nationality"] = "UK"
train.head()
```
